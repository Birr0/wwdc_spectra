```python
import pandas as pd
import glob
import os

# Define the directory where your SLURM job saved the CSVs
output_dir = './output'

# Find all the CSV files
csv_files = glob.glob(os.path.join(output_dir, 'counts_model_*_z*.csv'))

# List to hold the extracted data
data = []

for file in csv_files:
    df = pd.read_csv(file)
    
    # Since the counts are the same for all embed_types (orig, cond, uncond),
    # we can just grab the first row's data
    row = df.iloc[0]
    
    data.append({
        'z_max': float(row['z_lim']), 
        'Count': int(row['z_test_len']) # or s_test_len, they are the same
    })

# Convert to a DataFrame and sort by z_max ascending
df_counts = pd.DataFrame(data).sort_values('z_max')

# Construct the LaTeX string
latex_str = r"""\begin{table}[htpb]
    \centering
    \begin{tabular}{c|c}
        $z_{max}$ & \textbf{Count} \\
        \hline"""

for _, row in df_counts.iterrows():
    # Format z_max to exactly 3 decimal places (e.g., 0.100)
    z_str = f"{row['z_max']:.3f}"
    count_str = f"{int(row['Count'])}"
    
    latex_str += f"\n        {z_str} &  {count_str} \\\\"
    latex_str += "\n        \\hline"

latex_str += r"""
    \end{tabular}
    \caption{Number of galaxies in each cumulative redshift-limited sample after applying $M_r \leq -20.72$.}
    \label{tab:z_count}
\end{table}"""

# Print the result so you can copy-paste it into your LaTeX document
print(latex_str)
```
