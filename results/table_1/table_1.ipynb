```python
import pandas as pd
from pathlib import Path
```

# Linear

```python
# 1. Define the base directory where your folders are located.
base_dir = Path(".")

# 2. Find all CSV files in the relevant directories
csv_files = list(base_dir.rglob("linear_predictions_*_v2/*.csv"))

# 3. Read and combine all CSVs into a single DataFrame
dataframes = []
for file in csv_files:
    df = pd.read_csv(file)
    
    # Extract clean base model name from the folder (e.g., 'spender_I' or 'spender_II')
    clean_model_name = file.parent.name.replace("mlp_predictions_", "").replace("_v2", "")
    df['base_model'] = clean_model_name 
    
    dataframes.append(df)

# Concatenate all individual dataframes together
combined_df = pd.concat(dataframes, ignore_index=True)

# 4. Group by BOTH 'base_model' and 'embed_type' to stratify the results
summary_stats = combined_df.groupby(["base_model", "embed_type"])["r2"].agg(["mean", "std"]).reset_index()

# Rename columns for cleaner output
summary_stats.rename(columns={"mean": "r2_mean", "std": "r2_std"}, inplace=True)

print("\n--- Summary Statistics by Base Model ---")
print(summary_stats.to_string(index=False))
```
